```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

def manual_blend(img1: np.ndarray, img2: np.ndarray, alpha: float) -> np.ndarray:
    if img1.shape != img2.shape:
        img2 = cv2.resize(img2, (img1.shape[1], img1.shape[0]))
    f1 = img1.astype(np.float32)
    f2 = img2.astype(np.float32)
    blended_f = (1.0 - alpha) * f1 + alpha * f2
    return np.clip(blended_f, 0, 255).astype(np.uint8)

img1 = cv2.imread('./Image1.jpg')
img2 = cv2.imread('./Image2.jpg')
alpha = 0.5

blended = manual_blend(img1, img2, alpha)

cv2.imwrite('manual_blend.jpg', blended)

blended_rgb = cv2.cvtColor(blended, cv2.COLOR_BGR2RGB)
plt.figure(figsize=(8,6))
plt.imshow(blended_rgb)
plt.axis('off')
plt.title(f"Manual Blend α={alpha}")
plt.show()
```
